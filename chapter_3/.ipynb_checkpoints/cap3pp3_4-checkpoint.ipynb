#### Exemplo 3.4

```python
%matplotlib inline

import matplotlib
import numpy as np
from sympy import *
from sympy.plotting import plot

init_printing(use_unicode=True)
```

```python
# Indutância, [H]
theta, L0, L2, L4, i = symbols('theta L0 L2 L4 i')
L = L0 + L2*cos(2*theta) + L4*sin(4*theta)
```

```python
# Torque

T = ((i**2)/2)*diff(L,theta)
T
```

Avaliação numérica

```python
Tcond = T.subs({L0:0.0254, L2:0.0083, L4:0.0018, i:3.5})
```

```python
plot(Tcond,(theta, -pi, pi))
```

Para achar os máximos e mínimos locais, fazemos:

```python
maxminlocal = solve(diff(Tcond), theta)
maxminlocal
```

Ângulos:

```python
N(functions.elementary.complexes.re(maxminlocal[4]*180/(pi)))
```

```python
N(functions.elementary.complexes.re(maxminlocal[6]*180/(pi)))
```
