```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib
from sympy import *
init_printing(use_unicode=True)
from sympy.plotting import plot
import SchemDraw as schem
import SchemDraw.elements as e
#  :-)
#import matplotlib.pyplot as plt
#plt.xkcd()
```

Variáveis:
 
 A_0: parâmetro de área do entreferro da palheta[m²]  
 theta: ângulo da palheta [rad]  
 N: número de espiras []  
 i: corrente [A]
 g: entreferro [m]

```python
A_0, theta, N_e, i, g, mu0 = symbols('A_0 theta N_e i g mu0')
```

Expressão da área [m²] efetiva do entreferro em função do ângulo:

```python
A = A_0*(1 - (4*theta/pi)**2)
A
```

Relutância [1/H]:

```python
R = 2*g/(mu0*A)
R
```

Indutância [H]:

```python
L = (N_e**2)/R
L
```

Fluxo [Wb]:

```python
phi = N_e*i/R
phi
```

Enlace de fluxo [Wb]:

```python
λ = phi*N_e
λ
```

Para o caso em que $i = 5$A e $\theta = 0$, obtemos um valor para lambda:

```python
λ = λ.subs({A_0:((5/1000)*(1/1000)), N_e:450, i:5, g:0.9/1000, mu0:(4*pi*10**(-7)), theta:0})
λ
```

```python
# Energia [J]
W = (1/2)*( λ**2)/L
W
```

Avaliação numérica:

```python
W_plot = W.subs({A_0:((5/1000)*(1/1000)), N_e:450, i:5, g:0.9/1000, mu0:(4*pi*10**(-7))})
W_plot
```

Energia magnética armazenada em função de $\theta$, na região de operação $|\theta| \leq \frac{\pi}{6}$

```python
plot(W_plot,(theta,-pi/6,pi/6))
```
